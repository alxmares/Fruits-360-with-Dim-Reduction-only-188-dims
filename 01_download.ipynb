# **Cargar las imágenes y guardarlas en formapo .npy**
----

### **1. Training set**

```python
import os 
import numpy as np
from PIL import Image
 
# Configuración de la ruta
dataset_path = "D:/Fruit-Images-Dataset/Fruit-Images-Dataset-master/Training"

# Inicializar listas para almacenar imágenes y etiquetas
X_data = []
y_data = []

# Recorrer cada carpeta (clase) y cargar imágenes
for fruit_class in os.listdir(dataset_path):
    fruit_path = os.path.join(dataset_path, fruit_class)
    for image_name in os.listdir(fruit_path):
        image_path = os.path.join(fruit_path, image_name)
        image = Image.open(image_path)
        X_data.append(np.array(image))
        y_data.append(fruit_class)

# Convertir listas a arrays de NumPy
X_data = np.array(X_data)
y_data = np.array(y_data)
```

```python
# Guardar los arrays en archivos .npy
np.save('X_data.npy', X_data)
np.save('y_data.npy', y_data)

print("Imágenes y etiquetas guardadas en archivos .npy")
```

```python
import os
import numpy as np
from PIL import Image

# Configuración de la ruta
dataset_path = "D:/Fruit-Images-Dataset/Fruit-Images-Dataset-master/Training"

# Inicializar listas para almacenar imágenes y etiquetas
X_data = []
y_data = []

# Recorrer cada carpeta (clase) y cargar imágenes
for fruit_class in os.listdir(dataset_path):
    fruit_path = os.path.join(dataset_path, fruit_class)
    for image_name in os.listdir(fruit_path):
        image_path = os.path.join(fruit_path, image_name)
        
        # Abrir la imagen y convertirla a escala de grises
        image = Image.open(image_path).convert('L')
        
        # Convertir la imagen a un array de NumPy y agregar a la lista
        X_data.append(np.array(image))
        y_data.append(fruit_class)

# Convertir listas a arrays de NumPy
X_data = np.array(X_data)
y_data = np.array(y_data)

print("Imágenes cargadas en blanco y negro con shape:", X_data.shape)
```

```python
# Guardar los arrays en archivos .npy
np.save('X_data_bn.npy', X_data)
np.save('y_data_bn.npy', y_data)

print("Imágenes y etiquetas guardadas en archivos .npy")
```
